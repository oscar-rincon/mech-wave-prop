# Tarea 1: Mecánica del Medio Continuo


## Import required libraries

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
plt.rc('font', family='Arial', size=10) # Configurar la fuente en Arial 10
from matplotlib.ticker import MultipleLocator
import seaborn as sns 
```

```python
E1 = 200_000
E2 = 1_000_000
L = 1000
H = 10
t=np.arange(0,501,1)
sigma_x = 4 * t
u=(L/E1)*(sigma_x)
w=-(L*(H**2)/E2)*(sigma_x)
```

```python
# Graficar
plt.figure(figsize=(3.0, 3.0))
ax = plt.gca()
sns.lineplot(x=t, y=sigma_x, color="gray", linewidth=1, label="Data")
# Configurar las etiquetas y títulos
ax.set_xlabel(r'Tiempo (s)', fontsize=9)
ax.set_ylabel(r"$\sigma_{x}$", fontsize=9)
plt.xticks(np.arange(0,600,100),fontsize=9)
plt.yticks(np.arange(0,2200,200),fontsize=9)
plt.xlim([0,500])
plt.ylim([0,2000])
ax.xaxis.set_minor_locator(MultipleLocator(50))
ax.yaxis.set_minor_locator(MultipleLocator(100))
ax.legend().set_visible(False)
sns.despine()
ax.spines['bottom'].set_linewidth(0.7)  # Grosor del eje x
ax.spines['left'].set_linewidth(0.7)    # Grosor del eje y
ax.tick_params(axis='both', which='both', width=0.7)  # Ajustar width y length según sea necesario
# Mostrar el gráfico
plt.grid(False)
plt.tight_layout()
plt.savefig("sigma_x_t.pdf", format="pdf", dpi=300)
plt.show()
```

```python
# Graficar
plt.figure(figsize=(3.0, 3.0))
ax = plt.gca()
sns.lineplot(x=t, y=u, color="gray", linewidth=1, label="Data")
# Configurar las etiquetas y títulos
ax.set_xlabel(r'Tiempo (s)', fontsize=9)
ax.set_ylabel(r"$u$", fontsize=9)
#plt.xticks(np.arange(0,600,100),fontsize=9)
#plt.yticks(np.arange(0,2200,200),fontsize=9)
plt.xlim([0,500])
plt.ylim([0,11])
#ax.xaxis.set_minor_locator(MultipleLocator(50))
#ax.yaxis.set_minor_locator(MultipleLocator(100))
ax.legend().set_visible(False)
sns.despine()
ax.spines['bottom'].set_linewidth(0.7)  # Grosor del eje x
ax.spines['left'].set_linewidth(0.7)    # Grosor del eje y
ax.tick_params(axis='both', which='both', width=0.7)  # Ajustar width y length según sea necesario
# Mostrar el gráfico
plt.grid(False)
plt.tight_layout()
plt.savefig("u_t.pdf", format="pdf", dpi=300)
plt.show()
```

```python
# Graficar
plt.figure(figsize=(3.0, 3.0))
ax = plt.gca()
sns.lineplot(x=t, y=w, color="gray", linewidth=1, label="Data")
# Configurar las etiquetas y títulos
ax.set_xlabel(r'Tiempo (s)', fontsize=9)
ax.set_ylabel(r"$w$", fontsize=9)
#plt.xticks(np.arange(0,600,100),fontsize=9)
#plt.yticks(np.arange(0,2200,200),fontsize=9)
#plt.xlim([0,500])
#plt.ylim([0,11])
#ax.xaxis.set_minor_locator(MultipleLocator(50))
#ax.yaxis.set_minor_locator(MultipleLocator(100))
ax.legend().set_visible(False)
sns.despine()
ax.spines['bottom'].set_linewidth(0.7)  # Grosor del eje x
ax.spines['left'].set_linewidth(0.7)    # Grosor del eje y
ax.tick_params(axis='both', which='both', width=0.7)  # Ajustar width y length según sea necesario
# Mostrar el gráfico
plt.grid(False)
plt.tight_layout()
plt.savefig("w_t.pdf", format="pdf", dpi=300)
plt.show()
```
