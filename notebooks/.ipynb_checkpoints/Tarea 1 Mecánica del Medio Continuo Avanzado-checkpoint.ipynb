## Import required libraries

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'sans'
from matplotlib.ticker import MultipleLocator
import seaborn as sns 
%matplotlib inline
# use latex for font rendering
 
```

```python
E1 = 200_000
E2 = 1_000_000
L = 1000
H = 10
t=np.arange(0,501,1)
sigma_x = 4 * t
u=(L/E1)*(sigma_x)
w=-(L*(H**2)/E2)*(sigma_x)
```

```python
# Graficar
plt.figure(figsize=(3.0, 3.0))  # Increase figure size
ax = plt.gca()
sns.lineplot(x=t, y=sigma_x, color="blue", linewidth=1, label="Data")

# Configure labels and titles
ax.set_xlabel(r'Tiempo (s)', fontsize=10)  # Increase font size
ax.set_ylabel(r"$\sigma_{x}$", fontsize=10)  # Increase font size
plt.xticks(np.arange(0, 600, 100), fontsize=10)  # Increase font size
plt.yticks(np.arange(0, 2200, 200), fontsize=10)  # Increase font size
plt.xlim([0, 500])
plt.ylim([0, 2000])
ax.xaxis.set_minor_locator(MultipleLocator(50))
ax.yaxis.set_minor_locator(MultipleLocator(100))
ax.legend().set_visible(False)
sns.despine()
ax.spines['bottom'].set_linewidth(0.7)  # Adjust bottom spine width
ax.spines['left'].set_linewidth(0.7)  # Adjust left spine width
ax.tick_params(axis='both', which='both', width=0.7)  # Adjust tick parameters

# Include grid
plt.grid(True, which='both', linestyle='--', linewidth=0.5)  # Include grid lines

# Show the plot
plt.tight_layout()
plt.savefig("sigma_x_t.pdf", format="pdf", dpi=300)
plt.show()
```

```python
# Graficar
plt.figure(figsize=(3.0, 3.0))  # Increase figure size
ax = plt.gca()
sns.lineplot(x=t, y=u, color="blue", linewidth=1, label="Data")

# Configure labels and titles
ax.set_xlabel(r'Tiempo (s)', fontsize=10)  # Increase font size
ax.set_ylabel(r"$u(x)$ (cm)", fontsize=10)  # Increase font size
plt.xticks(np.arange(0, 700, 100), fontsize=10)  # Increase font size
plt.yticks(np.arange(0, 11, 1), fontsize=10)  # Increase font size
plt.xlim([0, 500])
plt.ylim([0, 10])
ax.legend().set_visible(False)
sns.despine()
ax.spines['bottom'].set_linewidth(0.7)  # Adjust bottom spine width
ax.spines['left'].set_linewidth(0.7)  # Adjust left spine width
ax.tick_params(axis='both', which='both', width=0.7)  # Adjust tick parameters
ax.xaxis.set_minor_locator(MultipleLocator(50))
ax.yaxis.set_minor_locator(MultipleLocator(0.5))
# Include grid
plt.grid(True, which='both', linestyle='--', linewidth=0.5)  # Include grid lines

# Show the plot
plt.tight_layout()
plt.savefig("u_t.pdf", format="pdf", dpi=300)
plt.show()
```

```python
plt.figure(figsize=(3.0, 3.0))  # Increase figure size
ax = plt.gca()
sns.lineplot(x=t, y=w, color="blue", linewidth=1, label="Data")

# Configure labels and titles
ax.set_xlabel(r'Tiempo (s)', fontsize=10)  # Increase font size
ax.set_ylabel(r"$w$ (cm)", fontsize=10)  # Increase font size
plt.xticks(np.arange(0, 700, 100), fontsize=10)  # Increase font size
plt.yticks(np.arange(-200, 10, 20), fontsize=10)  # Increase font size
plt.xlim([0, 500])
plt.ylim([-200, 1])
ax.xaxis.set_minor_locator(MultipleLocator(50))
ax.yaxis.set_minor_locator(MultipleLocator(10))
ax.legend().set_visible(False)
sns.despine()
ax.spines['bottom'].set_linewidth(0.7)  # Adjust bottom spine width
ax.spines['left'].set_linewidth(0.7)  # Adjust left spine width
ax.tick_params(axis='both', which='both', width=0.7)  # Adjust tick parameters

# Include grid
plt.grid(True, which='both', linestyle='--', linewidth=0.5)  # Include grid lines

# Show the plot
plt.tight_layout()
plt.savefig("w_t.pdf", format="pdf", dpi=300)
plt.show()
```
